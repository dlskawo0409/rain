```python
skew_features = ["WindGustSpeed","WindSpeed3pm","WindSpeed9am","Evaporation","Sunshine"]
scaled_MiceImputed_weather = weather_df.copy()
for feature in skew_features:
    amount_n = np.log1p(scaled_MiceImputed_weather[feature])
    scaled_MiceImputed_weather.insert(1, feature + "Scaled", amount_n)
    scaled_MiceImputed_weather.drop([feature], axis=1, inplace=True)
```

```python

plt.figure(figsize=(15,10))


plt.subplot(2, 2, 1)
fig = scaled_MiceImputed_weather.SunshineScaled.hist(bins=10)
fig.set_xlabel('Sunshine')



plt.subplot(2, 2, 2)
fig = scaled_MiceImputed_weather.EvaporationScaled.hist(bins=10)
fig.set_xlabel('Evaporation')



plt.subplot(2, 2, 3)
fig = scaled_MiceImputed_weather.WindSpeed9amScaled.hist(bins=10)
fig.set_xlabel('WindSpeed9am')



plt.subplot(2, 2, 4)
fig = scaled_MiceImputed_weather.WindSpeed3pmScaled.hist(bins=10)
fig.set_xlabel('WindSpeed3pm')
```
